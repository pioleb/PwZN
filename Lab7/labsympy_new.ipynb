```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
t = sp.Symbol('t')
x = sp.Function('x')
k = sp.Symbol('k')
b = sp.Symbol('b')

#initial conditions
x0 = sp.Symbol('x_0', real = True)
v0 = sp.Symbol('v_0', real = True)
```

```python
eq = sp.Eq(x(t).diff(t, t) + b*x(t).diff(t) + k*x(t), 0)
eq
```

```python
#sol = sp.dsolve(eq.subs({b: 0.5}))
sol = sp.dsolve(eq, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
```

```python
sol.simplify()
```

```python
sol.rhs.subs({x0: 1, v0: -1, b: 0.1})
```

```python
sp.plot(sol.rhs.subs({x0: 1, v0: -1, b: 0.1, k: 1}), (t, -20, 20))
```

```python
# b = 1
sp.plot(sol.rhs.subs({x0: 1, v0: -1, b: 1, k: 1}), (t, -20, 20))
```

```python
#b = 0.01
sp.plot(sol.rhs.subs({x0: 1, v0: -1, b: 0.01, k: 1}), (t, -20, 20))
```

```python
# z wymuiszeniem
a = sp.Symbol('a')
w = sp.Symbol('w')
eq2 = sp.Eq(x(t).diff(t, t) + x(t).diff(t) + k*x(t), a*sp.cos(w*t))
eq2
```

```python
sol2 = sp.dsolve(eq2, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol2
```

```python
sp.plot(sol2.rhs.subs({x0: 1, v0: -1, k: 1, a: 1, w: 1}), (t, -5, 5))
```

```python
sp.plot(sol2.rhs.subs({x0: 1, v0: -1, k: 1, a: 10, w: 0.5}), (t, -5, 5))
```

```python
#tłumienie i wymuszenie
eq3 = sp.Eq(x(t).diff(t, t) + b*x(t).diff(t) + k*x(t), a*sp.cos(w*t))
sol3 = sp.dsolve(eq3, x(t), ics = {x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol3
```

```python
sp.plot(sol3.rhs.subs({x0: 1, v0: -1, k: 1, b: 0.1, a: 1, w: 1}), (t, -20, 20))
```
